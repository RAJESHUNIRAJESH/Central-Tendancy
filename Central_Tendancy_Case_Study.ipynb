```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset file as Pandas DataFrame
a = pd.read_csv('/content/Wine Quality Dataset.csv')
data = a.copy()
data.head()   # take a breif look at the data
```

```python
# Get the row names of the dataframe
data.index
```

View the distributions of the various features in the dataset and calculate their central tendencies

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data,x = 'fixed acidity')
plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel("Count")
plt.show()

# normal distributed
```

```python
print(data['fixed acidity'].mean())
print(data['fixed acidity'].median())
```

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data,x = 'fixed acidity',bins = 5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel("Count")
plt.vlines(data['fixed acidity'].mean(),ymin = 0,ymax = 4000,color = 'blue',label = 'Mean')
plt.vlines(data['fixed acidity'].median(),ymin = 0,ymax = 4000,color = 'red',label = 'Median')
plt.show()
```

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data,x = 'volatile acidity')
plt.title("Histogram of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Count")
plt.show()

# right skewed
```

```python
plt.figure(figsize = (11,6))
sns.distplot(data['volatile acidity'])
plt.title("Distplot of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Count")
plt.show()
```

```python
print(data['volatile acidity'].skew())
print(data['volatile acidity'].mean())
print(data['volatile acidity'].median())
```

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data,x = 'volatile acidity',bins = 5)
plt.title("Histogram of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Count")
plt.vlines(data['volatile acidity'].mean(),ymin = 0,ymax = 4000,color = 'blue',label = 'Mean')
plt.vlines(data['volatile acidity'].median(),ymin = 0,ymax = 4000,color = 'red',label = 'Median')
plt.show()
```

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data,x = 'citric acid')
plt.title("Histogram of Volatile Acidity")
plt.xlabel("Citric Acid")
plt.ylabel("Count")
plt.show()
```

```python
print(data['citric acid'].skew())
print(data['citric acid'].mean())
print(data['citric acid'].median())
```

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data,x = 'citric acid',bins = 5)
plt.title("Histogram of Volatile Acidity")
plt.xlabel("Citric Acid")
plt.ylabel("Count")
plt.vlines(data['citric acid'].mean(),ymin = 0,ymax = 4000,color = 'blue',label = 'Mean')
plt.vlines(data['citric acid'].median(),ymin = 0,ymax = 4000,color = 'red',label = 'Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (11,6))
sns.distplot(data['citric acid'])
plt.title("Histogram of Volatile Acidity")
plt.xlabel("Citric Acid")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (11,6))
sns.countplot(data['quality'])
plt.title("Count plot of Quality")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```
